```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("population.csv")
```

```python
top10 = df.head(10)
top10
```

```python
ct_name = top10["Country (or dependency)"]
top10_pp = top10["Population (2020)"]
```

```python
title_font = {'family':'serif','color':'blue','size':20}
label_font = {'family':'serif','color':'darkred','size':15}
```

#### 10 đất nước có diện tích lớn nhất thế giới

```python
area_df = df.copy()
area_df.sort_values("Land Area (Km²)", axis = 0,ascending = False, inplace = True)
area_df = area_df.head(10)
top_area = area_df["Land Area (Km²)"]
area_name = area_df["Country (or dependency)"]
```

```python
plt.figure(figsize=(16, 10))
bars = plt.barh(area_name,top_area)
# To get data labels
ax1 = plt.subplot(1,1,1)
for y,x in enumerate(top_area):
  ax1.annotate("{:,}".format(x), xy=(x,y), va="center")

plt.title("Top 10 largest land area countries", fontdict = title_font)
plt.xlabel("Country name", fontdict = label_font)
plt.ylabel("Land area ", fontdict = label_font)
# Display a bar chart
plt.show()
```
